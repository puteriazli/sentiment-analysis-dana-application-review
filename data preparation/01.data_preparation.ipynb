# Data Modification

```python
kolom_dibuang = ['userImage', 'reviewCreatedVersion', 'replyContent', 'repliedAt']
df.drop(columns=kolom_dibuang, inplace=True, errors='ignore')
```

```python
df.rename(columns={
    'reviewId': 'id_ulasan',
    'userName': 'user',
    'content': 'ulasan',
    'score': 'rating',
    'thumbsUpCount': 'dukungan_ulasan',
    'at': 'tanggal',
    'appVersion': 'versi_aplikasi'
}, inplace=True)

df.info()
```

```python
df['tanggal'] = pd.to_datetime(df['tanggal'])
df['versi_aplikasi'] = df['versi_aplikasi'].astype('category')
print(df.info(memory_usage='deep'))
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
print(df[df['ulasan'].isna()])
```

```python
df['ulasan'] = df['ulasan'].fillna('no review')
print(f"Jumlah missing values sekarang: {df['ulasan'].isna().sum()}")
print("\nCek ulasan yang baru diisi:")
df[df['ulasan'] == 'no review']
```

```python
df['versi_aplikasi'] = df['versi_aplikasi'].cat.add_categories(['unknown']).fillna('unknown')
df[df['versi_aplikasi'] == 'unknown']
```
