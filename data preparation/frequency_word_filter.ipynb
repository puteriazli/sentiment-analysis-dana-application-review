# Data Hasil Shorted

```python
df = pd.read_csv('../frequency_dictionary_id_sorted.csv')
df.head(10)
```

```python
df["kata_final"] = df["koreksi"].replace("", pd.NA).fillna(df["kata"])
```

```python
df.head(20)
```

```python
# pakai hanya kolom yang dibutuhkan
df_final = df[["kata_final", "frekuensi"]].copy()
# rename kolom
df_final = df_final.rename(columns={"kata_final": "kata"})

# simpan ke file baru
df_final.to_csv("../data/kamus/frequency_dictionary_cleaned.csv", index=False)
```
