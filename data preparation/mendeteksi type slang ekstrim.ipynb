```python
# cek fregmentasi
from transformers import AutoTokenizer

MODEL_NAME = "indobenchmark/indobert-base-p1"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

def analyze_fragmentation(text):
    words = text.split()
    results = []
    
    for w in words:
        tokens = tokenizer.tokenize(w)
        results.append((w, tokens, len(tokens)))
        
    return results

sample = "aplikasinya bgusszzzz bangettt"
print(analyze_fragmentation(sample))
```
